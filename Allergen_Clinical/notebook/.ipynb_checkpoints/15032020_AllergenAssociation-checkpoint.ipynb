```python
qbb_spp = pd.read_pickle("../../QBBData/Data_PhipSeq_BWH/merge_800_10Feb2020/species_score.pkl",compression='gzip')
virCut = pd.read_csv("../../QbbBMI_Project/curatedData/VirscanThreshold_3Feb2020.csv").set_index('Species')
qbbmData = pd.read_pickle("../../QBBData/MetaData/QBB_GAIBNE_16032020_cluster.pkl")#.set_index('PROJECT_DUMMY_ID')
hla_gt = pd.read_pickle("../../QBBData/Notebook/curatedFiles/HLA_genotype.pkl")

print (virCut[(virCut.GLM_2ft>=2)&(virCut.prev_qbb>1)].shape)
spp_posFrac = qbb_spp.loc[virCut[virCut.GLM_2ft>=2].index].divide(virCut[virCut.GLM_2ft>=2]['GLM_2ft'],axis=0)
spp_posBool = spp_posFrac[spp_posFrac>=1].fillna(0).astype('bool')
```

```python
_ = alerData>=0.3
at_oneAll = _.loc[_.sum(axis=1)>0].index
_colid = set(spp_posFrac.columns).intersection(set(at_oneAll))
spp_posFrac[_colid].shape
```

```python
_frac = spp_posBool[_colid].sum(axis=1)/len(_colid)
sp_alg5 = _frac[_frac>0.05].index
print (len(sp_alg5))
```

```python
spfrac = spp_posFrac.loc[sp_alg5][_colid]
spbool = spp_posBool.loc[sp_alg5][_colid]
print (spfrac.shape, spbool.shape)
```

```python
# the selected 60 species have occurance frequency greater than 5% and species specific cutoff
# score atleast 2.
```

```python
X = spfrac.transpose().values
y = alerData.loc[spfrac.columns].phad.values
```

```python
print (X.shape, y.shape)
```

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(y,X)
results = model.fit()
tdata = pd.read_html(results.summary().tables[1].as_html(),header=0,index_col=0)[0]
tdata['species'] = spfrac.index
tdata[tdata["P>|t|"]<=0.1]
```

```python
tdata.shape
```

```python
print (X.shape, y.shape)
```

```python
y_bool = alerData.loc[spfrac.columns].phad>=0.3
```

```python
model_logit = sm.Logit(y_bool.values,X)
results_logit = model_logit.fit()
tdata_lg = pd.read_html(results_logit.summary().tables[1].as_html(),header=0,index_col=0)[0]
tdata_lg['species'] = spfrac.index
tdata_lg[tdata_lg["P>|z|"]<=0.1]
```

```python
X = spfrac.transpose().values
# y = alerData.loc[spfrac.columns].phad.values

aller_ols_vir = pd.DataFrame()
for atype in alerData.columns:
    y_ = alerData.loc[spfrac.columns][atype].values
    model = sm.OLS(y_,X)
    results = model.fit()
    tdata = pd.read_html(results.summary().tables[1].as_html(),header=0,index_col=0)[0]
    tdata['species'] = spfrac.index
    tdata['testType'] = atype
    _signi = tdata[tdata["P>|t|"]<=0.1]
    aller_ols_vir = pd.concat([aller_ols_vir,_signi],axis=0)
    
vir_coeff_ols = aller_ols_vir.groupby(['species','testType'])['coef'].apply(float).unstack()
```

```python

aller_logit_vir = pd.DataFrame()
for atype in alerData.columns:
    y_bool = alerData.loc[spfrac.columns][atype]>=0.3
    model_logit = sm.Logit(y_bool.values,X)
    results_logit = model_logit.fit()
    tdata_lg = pd.read_html(results_logit.summary().tables[1].as_html(),header=0,index_col=0)[0]
    tdata_lg['species'] = spfrac.index
    tdata_lg['testType'] = atype
    _signi_lg = tdata_lg[tdata_lg["P>|z|"]<=0.1]
    aller_logit_vir = pd.concat([aller_logit_vir,_signi_lg],axis=0)
    
vir_coeff_logit = aller_logit_vir.groupby(['species','testType'])['coef'].apply(float).unstack()
```

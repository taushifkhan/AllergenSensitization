```python
import pandas as pd
import numpy as np
```

```python
def to_pdframe(fid):
    f = open(fid,'r').readlines()
    print ("reading", fid)
    dm = 0
    for i in np.arange(len(f)):
        if f[i].strip() == "[MEAN]":
            dm=i
            break
    dk = f[i+1:len(f)]
    l = []
    for k in dk[1:]:
        l.append((k.strip().split(",")))
    _df = pd.DataFrame(l,columns=[i.strip("(|)") for i in dk[0].split(",")])
    print ("DF:", _df.shape)
    return _df
```

```python
from glob import glob
```

```python
pd_ige = pd.DataFrame()
for i in glob("../Allergen data/*.csv"):
    pd_ige = pd.concat([pd_ige,to_pdframe(i)],axis=0)
```

```python
sv = pd_ige[pd_ige.SampleType.str.contains("SAM")].Sample.value_counts()
pd_qbb = pd_ige[pd_ige.Sample.isin(sv[sv==8].index)][['Sample','Test','Conc']]
```

```python
pd_ige.head()
```

```python
pd_qbb['Conc'] = pd_qbb['Conc'].astype('float')
```

```python
_x = pd_qbb.groupby(['Test','Sample'])['Conc'].agg('mean').unstack().fillna(0)
```

```python
import seaborn as sns
```

```python
test_cr = _x.loc[_x[_x>0].fillna(0).astype(bool).sum(axis=1)>1]
```

```python
test_cr[test_cr>0].fillna(0).astype(bool).sum().plot(kind="hist")
plt.xlabel("Numbr of Test",fontsize=14)
plt.ylabel("sample Frequency",fontsize=14)
```

```python
test_cr.columns[test_cr[test_cr>0].fillna(0).astype(bool).sum()>=5]
```

```python
s_col = test_cr.columns[test_cr[test_cr>0].fillna(0).astype(bool).sum()>=5]
sns.clustermap(test_cr[s_col].transpose(),vmin=0.01, vmax=1,cmap="Blues",figsize=(5,16))
plt.yticks(s_col,[])
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(test_cr[s_col].transpose().corr(),cmap="RdPu",annot=True)
```

```python
samplesIgElevels = []
for i in np.arange(0,1.1,0.1):
    i = round(i,2)
    print (i)
    for kname,kgrp in igE_rawX.groupby(['Allergen']):
        _pscore = kgrp.IgEscore.quantile(i)
        _samples = kgrp[kgrp.IgEscore>=_pscore].shape[0]
        samplesIgElevels.append([kname,i,_pscore,_samples])
        
samplesIgElevels = pd.DataFrame(samplesIgElevels,columns=['panel','Q_val','IgEScore_cutoff','sampleCount'])
```

```python

for kname, kgrp in samplesIgElevels.groupby("panel"):
    _igCUtoff = kgrp[(kgrp.IgEScore_cutoff>=0.2)&(kgrp.IgEScore_cutoff<=0.6)]
    print (kname, _igCUtoff)
```

```python
samplesIgElevelsX = samplesIgElevels[(samplesIgElevels.IgEScore_cutoff>0.1)&\
                 (samplesIgElevels.IgEScore_cutoff<0.8)].sort_values('panel')
```

```python
IgE['QBB_DUMMY_ID'] = [i.split("-")[0][-15:] for i in IgE['sample'].values]
IgETotal = IgE[['QBB_DUMMY_ID','QBB',"IVIg","IgG depleted"]].set_index("QBB_DUMMY_ID")
```

```python
IgETotal.dropna()
```

```python
IgETotal.reset_index().drop_duplicates(subset="QBB_DUMMY_ID")
```

```python
_commonIDs = set(igE_clinic.index).intersection(set(IgETotal.index))
igE_clinic_Sum = pd.DataFrame(index=_commonIDs)
igE_clinic_Sum['totalPos'] = igE_clinic.loc[_commonIDs].sum(axis=1)
```

```python
igE_clinic_Sum#['TotalIgE'] = IgETotal.loc[_commonIDs]["QBB"].values
```

```python
IgEDF = pd.concat([pd.DataFrame(IgETotal.loc[igE_clinic_Sum.index].QBB).reset_index().drop_duplicates(subset="QBB_DUMMY_ID").set_index("QBB_DUMMY_ID"),igE_clinic_Sum],axis=1).rename({"QBB":'totalIgE_ngPml'},axis=1)
```

```python
# convert ng/ml to IU/ml for IgE 1 ng/ml = 0.4167 IU/ml
IgEDF['totalIgE_mgPml'] = IgEDF.totalIgE_ngPml*0.4167
```

```python
IgEDF
```

```python
sns.boxplot(x='totalPos',y='totalIgE_mgPml',data=IgEDF)
sns.swarmplot(x='totalPos',y='totalIgE_mgPml',data=IgEDF,color='k',alpha=0.2)
```

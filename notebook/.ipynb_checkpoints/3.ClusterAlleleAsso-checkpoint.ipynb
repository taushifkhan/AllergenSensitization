```python
def posNegAsso(kgrp):
    pos = kgrp[kgrp.meanCoeff>0].shape[0]
    neg = kgrp[kgrp.meanCoeff<0].shape[0]
    total = pos+neg
    total2 = kgrp.shape[0]
    return pd.Series({'pos':pos,'neg':neg,'totalAsso':total,'total2':total2,'sigCount':"{} ({}/{})".format(total,pos,neg)})

coefs_c1_sigCount = coefs_c1.groupby("allele").apply(posNegAsso).join(alleleFrqSel)
coefs_c1_sigCount
```

```python
from Bio import Entrez
def search(query):
    Entrez.email = 'your.email@example.com'
    handle = Entrez.esearch(db='pubmed',
                            sort='relevance',
                            retmax='20',
                            retmode='xml',
                            term=query)
    results = Entrez.read(handle)
    return results

def fetch_details(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'tkhan@sidra.org'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    results = Entrez.read(handle)
    return results
```

```python
searchIdx =pd.DataFrame(index=newRefTable[newRefTable.isna().sum(axis=1)!=0].index)
for k in searchIdx.index:
    query = "[{}] AND Allergy".format(k.replace("*","\*"))
    results = search(query)
    if int(results['Count'])>=1:
        searchIdx.loc[k,'count'] = int(results['Count'])
        searchIdx.loc[k,'pubmedID'] = ",".join(results['IdList'])
        
# sigAlleleAllergy = searchIdx.loc[tx.allele.unique()]

pubMedList = []
pubMedListKeys = {}

for i in searchIdx.dropna().index:
    pubMedList.extend(searchIdx.loc[i,'pubmedID'].split(","))
    for x in searchIdx.loc[i,'pubmedID'].split(","):
        if x in list(pubMedListKeys.keys()):
            pubMedListKeys[x].append(i)
        else:
            pubMedListKeys[x]= [i]
            
pubMedList = set(pubMedList)
print (len(pubMedList),len(list(pubMedListKeys.keys())))
```

### Examine associaiton from Model 1 and Model 2 together and project with allele frequency

```python
m1DF_Curated_prgXAnnot = pd.read_csv("../tables/Monosensitization_sigAnnot.csv").set_index("test")
m2DF_Curated_aggJ = pd.read_csv("../tables/model2_aggAssociations.csv").set_index("test")
set(m2DF_Curated_aggJ.index).intersection(set(m1DF_Curated_prgXAnnot.index))
```

```python
monoMultiDFGrp_Tab= pd.read_csv("../tables/multi-monoAssociationValues.csv")
```

```python
alleleAllergenFreq = pd.read_csv("../data_curated/Allele-AllergenFrequency.csv").set_index("allele_allergen")
alleleAllergenFreq.head()
```

```python
alleleFrq = pd.read_csv("../data_primary/AlleleFrequency_protRes_2methods.csv").set_index("ProtRes")
print (alleleFrq.shape)
alleleFrqSel = alleleFrq[alleleFrq.selectedForStudy==1]
alleleFrqSel['mAF'] = alleleFrqSel[['AFprot_hd','AFprot_prg']].mean(axis=1)
alleleFrqSel =alleleFrqSel[['mAF','gene','molecule']]
print (alleleFrqSel.shape)
alleleFrqSel
```

```python
alleleAllergenFreq['OR_la'] = np.log(alleleAllergenFreq['posFrac_LA']/alleleAllergenFreq['negFrac_LA'])
alleleAllergenFreq['OR_hd'] = np.log(alleleAllergenFreq['posFrac_HD']/alleleAllergenFreq['negFrac_HD'])
```

```python
mono_multi = set(m2DF_Curated_aggJ.index).union(set(m1DF_Curated_prgXAnnot.index))
print (len(mono_multi))
alleleAllergenFreqI1 = alleleAllergenFreq.loc[mono_multi]
```

```python
alleleAllergenFreqI1["OR_combined"] = alleleAllergenFreqI1[['OR_la','OR_hd']].mean(axis=1)
```

```python
alleleAllergenFreqI1.head()
```

```python
chopCombo = alleleAllergenFreqI1.join(monoMultiDFGrp_Tab.set_index("AlleleAllergen"))
```

```python
iters = chopCombo[chopCombo.columns[chopCombo.columns.str.contains("iter")]].fillna(0)
coefs = chopCombo[chopCombo.columns[chopCombo.columns.str.contains("coef")]].fillna(0)
```

```python
class1_allSig = iters[iters.min(axis=1)>=90]
class1_allSig
```

```python
coefs_c1 = coefs.loc[class1_allSig.index].median(axis=1).to_frame().rename({0:'meanCoeff'},axis=1)
coefs_c1['allele'] = [i.split("_")[1] for i in coefs_c1.index]
coefs_c1['allergy'] = [i.split("_")[0] for i in coefs_c1.index]
```

```python
coefs_c1.join(chopCombo['OR_combined'])
```

```python
def posNegAsso(kgrp):
    return pd.Series({'pos':kgrp[kgrp.meanCoeff>0].shape[0],'neg':kgrp[kgrp.meanCoeff<0].shape[0],'totalAsso':kgrp.shape[0]})
coefs_c1.groupby("allele").apply(posNegAsso).join(alleleFrqSel)
```

```python
coefs_c1[coefs_c1.allele=="DPB1*09:01"][['allergy','meanCoeff']].set_index("allergy").to_dict()
```

```python
class2_allSig = iters[iters.min(axis=1)<90]
print (class2_allSig.shape)

class2_allSig = coefs.loc[class2_allSig.index].median(axis=1).to_frame().rename({0:'meanCoeff'},axis=1)
class2_allSig['allele'] = [i.split("_")[1] for i in class2_allSig.index]
class2_allSig['allergy'] = [i.split("_")[0] for i in class2_allSig.index]

class2_allSig.groupby("allele").apply(posNegAsso).join(alleleFrqSel)
```

```python
### any associaiton
coefsAll = coefs.median(axis=1).to_frame().rename({0:'meanCoeff'},axis=1)
coefsAll['allele'] = [i.split("_")[1] for i in coefsAll.index]
coefsAll['allergy'] = [i.split("_")[0] for i in coefsAll.index]

coefsAllAF = coefsAll.groupby("allele").apply(posNegAsso).join(alleleFrqSel)
```

```python
associationCountx = pd.concat([coefsAllAF, alleleFrqSel.loc[~alleleFrqSel.index.isin(coefsAllAF.index)]],axis=0).sort_values(by=['totalAsso','mAF','gene']).fillna(0)
```

```python
associationCountx.to_csv("../tables/associaitonAlleleFreq.csv")
```
